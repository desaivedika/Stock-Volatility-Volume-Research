```python
import matplotlib.pyplot as plt
import numpy as np

assets = ["AAPL", "GOOG", "ABNB", "BTC", "ETH"]
poly_r2 = [0.2165, 0.2541, 0.1315, 0.2763, 0.4327]

plt.figure(figsize=(10, 6))
bars = plt.bar(assets, poly_r2, color=["#4c72b0", "#55a868", "#c44e52", "#8172b2", "#ccb974"])

plt.title("Polynomial Model R² Comparison Across All Assets", fontsize=16, fontweight="bold")
plt.xlabel("Asset", fontsize=13)
plt.ylabel("R² Value", fontsize=13)
plt.grid(axis='y', alpha=0.3)

# Add values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01,
             f"{yval:.3f}", ha='center', fontsize=12)

plt.show()
```

Final discussion
The analysis shows that the relationship between volume changes (ΔV) and volatility changes (Δσ) exists but varies significantly across assets. Among the three stocks, GOOG showed the strongest fit (R² ≈ 0.25), while ABNB had the weakest (R² ≈ 0.13), indicating that stock-specific factors influence how volume affects volatility. AAPL demonstrated a moderate relationship with an R² of about 0.21.
Cryptocurrencies displayed a noticeably stronger connection. BTC achieved an R² of around 0.27—higher than most stocks—while ETH showed the highest overall explanatory power (R² ≈ 0.43), suggesting that crypto volatility is more sensitive to changes in trading volume. This may be due to lower market liquidity and higher speculative activity.
Polynomial models did not significantly outperform linear models for most assets, except for small improvements in GOOG and ETH. Overall, the findings indicate that while volume contributes to explaining volatility, it is not the dominant factor, and the strength of the relationship depends heavily on the asset type.




Summary
Across five assets—AAPL, GOOG, ABNB, BTC, and ETH—I modeled the relationship between trading volume changes and volatility changes using linear and polynomial regression. The analysis shows that the relationship exists but is generally weak for most stocks, with R² values around 0.13–0.25. Cryptocurrencies demonstrated stronger relationships, particularly ETH, which achieved the highest R² (≈0.43). Polynomial models offered marginal improvements only for GOOG and ETH. Overall, while volume provides some insight into volatility behavior, additional factors—such as market sentiment, liquidity, and macroeconomic conditions—significantly influence daily price movements.
